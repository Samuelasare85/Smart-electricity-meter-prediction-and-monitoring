```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
import statistics as stats
import mplcursors
import plotly.express as px
import datetime
from datetime import datetime
```

```python
os.listdir()
```

```python
smart_meter_data = pd.read_csv('log.csv',usecols=['balance','lastTopupAmount','date'])
```

```python
smart_meter_data['date'] = pd.to_datetime(smart_meter_data['date'])
smart_meter_data.head(5)
```

```python
smart_meter_data.loc[0, 'date'].day_name()
```

```python
new_format = "%Y-%m-%d"
smart_meter_data.loc[0, 'date'].strftime(new_format)
```

```python
balance = smart_meter_data['balance'][26:55]
date_time = smart_meter_data['date'][26:55]
```

```python
%matplotlib inline 
```

```python
plt.figure(figsize=(20,10))
Lines_plot = plt.plot(date_time, balance)
scatter_plot = plt.scatter(date_time, balance)
plt.xticks(rotation='vertical', size=15)
plt.xlabel('Date', fontsize=18)
plt.ylabel('Balance', fontsize=18)
plt.title('Plot of balance against time', fontsize=18)
plt.grid()
# plt.savefig('Plot.png')


```

# Finding the average 

```python
mean_balance = balance.diff()
mean_balance = abs(mean_balance)
mean_balance = mean_balance.dropna()
mean_balance = stats.mean(mean_balance)

print('Your daily average consumption is GH₵%s'%round(mean_balance,2))
```
